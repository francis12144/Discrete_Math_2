Question 1: Find a general formula for the expected value of this distribution.


Step 1:
Using the concept of the expected value for discrete random variables, we can get the expected value of the given probability distribution, which is a Poisson distribution with parameter $\lambda$, represented as $E(Y)$:

$$
E(Y) = \sum_{y=0}^{\infty} y \cdot p(y)
$$

Step 2:
Substituting the expression for $p(y)$ given in the problem, we get:

$$
E(Y) = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}
$$

Step 3:
We'll utilize the knowledge that a probability distribution's total probability is equal to $1$ to solve this. As a result, the total may be rewritten as follows:

$$
E(Y) = \sum_{y=1}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} + 0 \cdot \frac{\lambda^0}{0!} e^{-\lambda}
$$

Step 4:
Now, we can simplify the terms with $y$ greater than $0$:

$$
E(Y) = \sum_{y=1}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}
$$

Step 5:
To simplify further, let's take $e^{-\lambda}$ outside the sum:

$$
E(Y) = e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!}
$$

Step 6:
Let's now substitute anything to make the expression inside the sum simpler. Assuming $k = y - 1$, $y = k + 1$. Moreover, observe that $k = 0$ when $y = 1$, and that $k$ approaches infinity as well as $y$ approaches infinity. After making these changes, the total becomes:

$$
E(Y) = e^{-\lambda} \sum_{k=0}^{\infty} \frac{\lambda^{k+1}}{k!}
$$

$$
E(Y) = e^{-\lambda} \lambda \sum_{k=0}^{\infty} \frac{\lambda^k}{k!}
$$

Step 7:
The sum inside the expression is recognizable as the Taylor series expansion of $e^\lambda$, which is $e^\lambda$. Therefore:

$$
E(Y) = e^{-\lambda} \lambda e^\lambda
$$

$$
E(Y) = \lambda
$$

Step 8:
Thus, $\lambda$ is the universal formula for the expected value of the given distribution.

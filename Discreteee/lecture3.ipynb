$\textbf{17. Travelling Salesman problem}$
-

```python
import matplotlib.pyplot as plt
import networkx as nx

node_names = ["A", "B", "C", "D"]
edges = [('A', 'B', {'weight': 30}),
         ('A', 'C', {'weight': 50}),
         ('A', 'D', {'weight': 40}),
         ('B', 'C', {'weight': 30}),
         ('B', 'D', {'weight': 35}),
         ('C', 'D', {'weight': 25})]

routes = [
    ['A', 'B', 'C', 'D', 'A'],
    ['A', 'B', 'D', 'C', 'A'],
    ['A', 'C', 'B', 'D', 'A'],
    ['A', 'C', 'D', 'B', 'A'],
    ['A', 'D', 'B', 'C', 'A']
]

fig, axs = plt.subplots(1, 5, figsize=(20, 4))

for i, ax in enumerate(axs):
    G = nx.Graph()
    G.add_nodes_from(node_names)
    G.add_edges_from(edges)
    pos = nx.planar_layout(G)

    nx.draw_networkx_edge_labels(
        G, pos,
        edge_labels={edge[:2]: str(edge[2]['weight']) for edge in edges},
        font_color='red', ax=ax
    )

    nx.draw(G, pos=pos, with_labels=True, node_size=700, ax=ax)

    # Highlighting the specified route
    for j in range(len(routes[i]) - 1):
        nx.draw_networkx_edges(
            G, pos, edgelist=[(routes[i][j], routes[i][j + 1])],
            edge_color='cyan', width=2, ax=ax
        )

    # Calculating the weight of the route
    weight = sum(G[routes[i][j]][routes[i][j + 1]]['weight'] for j in range(len(routes[i]) - 1))
    ax.set_title('Route {}'.format(i + 1))
    ax.text(0.5, -0.15, 'Weight: {}'.format(weight), horizontalalignment='center', verticalalignment='center',
            transform=ax.transAxes)

# Adjust layout
plt.tight_layout()
plt.gcf().canvas.manager.set_window_title("TRAVELLING SALESMAN PROBLEM")
plt.show()
```

As you can see from the five possible routes of the graph. The best possible and shortest path are:

Route 1: $A \to B \to C \to D \to A$.



```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the nodes and edges
nodes = ['A', 'B', 'C', 'D', 'E', 'F']

edge_labels = {
    ('A', 'B'): 10,
    ('A', 'C'): 20,
    ('A', 'D'): 30,
    ('A', 'E'): 40,
    ('A', 'F'): 50,
    ('B', 'C'): 60,
    ('B', 'D'): 70,
    ('B', 'E'): 80,
    ('B', 'F'): 90,
    ('C', 'D'): 100,
    ('C', 'E'): 110,
    ('C', 'F'): 120,
    ('D', 'E'): 130,
    ('D', 'F'): 140,
    ('E', 'F'): 150,
}

# Convert edge labels to the desired format
formatted_edge_labels = {edge: "{}".format(weight) for edge, weight in edge_labels.items()}

# Define Hamilton circuits
hamilton_circuits = [
    ['A', 'B', 'C', 'D', 'E', 'F', 'A'],
    ['A', 'B', 'D', 'F', 'C', 'E', 'A'],
    ['A', 'C', 'E', 'F', 'B', 'D', 'A'],
    ['A', 'D', 'F', 'C', 'B', 'E', 'A'],
    ['A', 'E', 'F', 'D', 'C', 'B', 'A']
]

# Plotting
fig, axs = plt.subplots(1, 5, figsize=(20, 4))

for i in range(5):
    G = nx.Graph()
    G.add_nodes_from(nodes)

    # Add edges with formatted labels
    for edge, label in formatted_edge_labels.items():
        G.add_edge(edge[0], edge[1], weight=edge_labels[edge])

    pos = nx.circular_layout(G)

    nx.draw_networkx_edge_labels(
        G, pos,
        edge_labels=formatted_edge_labels,
        font_color='red', ax=axs[i]
    )

    nx.draw(G, pos=pos, with_labels=True, node_size=700, ax=axs[i])

    # Highlighting the specified Hamilton circuit
    for j in range(len(hamilton_circuits[i]) - 1):
        nx.draw_networkx_edges(
            G, pos, edgelist=[(hamilton_circuits[i][j], hamilton_circuits[i][j + 1])],
            edge_color='cyan', width=2, ax=axs[i]
        )

    # Calculating and displaying the weight of the route
    weight = sum(G[hamilton_circuits[i][j]][hamilton_circuits[i][j + 1]]['weight'] for j in range(len(hamilton_circuits[i]) - 1))
    axs[i].set_title('Route {}'.format(i + 1))
    axs[i].text(0.5, -0.15, 'Weight: {}'.format(weight), horizontalalignment='center', verticalalignment='center',
                transform=axs[i].transAxes)

# Adjust layout
plt.tight_layout()
plt.show()
```

As you can see from the five possible routes of the graph. The best possible and shortest path are:

Route 3: $A \to C \to E \to F \to B \to D \to A$.
